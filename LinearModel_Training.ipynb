```python
import numpy as np
```

```python
class Linear_Regression:

   # initiating the parameters
  def __init__(self,learning_rate,no_of_iterations):

       self.learning_rate=learning_rate
       self.no_of_iterations=no_of_iterations

  def fit(self,X,Y):
      # number of training examples and number of features

        self.m , self.n=X.shape # no of rows and Column from (X/input features)

      # initializing the weight and bias
        self.w=np.zeros(self.n)
        self.b=0
        self.X=X
        self.Y=Y

  # Implementing Gradient Descent

        for i in range(self.no_of_iterations):
           self.update_weights()

  def update_weights(self, ):

       Y_prediction=self.predict(self.X)

    # calculate gradients

       dw= -(2*(self.X.T).dot(self.Y-Y_prediction)) / self.m
       db=  -2 * np.sum(self.Y- Y_prediction)/self.m

    # updating the weights

       self.w= self.w - self.learning_rate*dw
       self.b = self.b - self.learning_rate*db




  def predict(self , X):

     return X.dot(self.w)+self.b

```

```python
# importing the depencies
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# loading the data from csv file to a pandas dataframe
salary_data = pd.read_csv('/content/salary_datasetR.csv')
```

```python
salary_data.head()
```

```python
# no of rows and column in dataFrame
salary_data.shape
```

```python
# checking for missing value
salary_data.isnull().sum()
```

Splitting The Feature And Target


```python
X=salary_data.iloc[:,:-1].values  # all row and all column except last as it is target
Y=salary_data.iloc[:,1].values
```

```python
print(X)
```

```python
print(Y)
```

 Splitting the datset into training and testing data

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.33, random_state=2)
```

Training the Linear Regression Model

```python
model= Linear_Regression(learning_rate=0.01,no_of_iterations=1000)
```

```python
model.fit(X_train , Y_train)
```

```python
# printing the parameter values(w and b)
print('weight = ', model.w[0])
print('bias = ' , model.b)
```

```python
test_data_prediction= model.predict(X_test)
```

```python
print(test_data_prediction)
```

Visualizing the predicted and Actual value

```python
plt.scatter(X_test, Y_test,color='red')
plt.plot(X_test, test_data_prediction , color='blue')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary Vs Experience')
plt.show()
```
